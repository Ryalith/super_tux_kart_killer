```python
def get_observation_vector_dim(env_spec):
    """
    Calculate the dimension of the flattened observation vector after processing.
    - Continuous observations: kept as-is
    - Discrete observations (MultiOneHot): already one-hot encoded, kept as-is

    Args:
        env_spec: Environment spec (from env.specs)

    Returns:
        int: Total dimension of the processed observation vector
    """
    # Handle if passed full env spec or just observation spec
    obs_spec = env_spec["output_spec"]["full_observation_spec"]
    total_dim = 0

    # Handle composite observation spaces
    for key in obs_spec.keys():
        spec = obs_spec[key]
        # Each component contributes its flattened shape
        print(f"{key=}")
        print(f"{spec=}")
        print(f"{spec.shape=}")
        total_dim += spec.shape.numel()

    return total_dim

get_observation_vector_dim(env.specs)
```

```python
def get_action_vector_dims(env_specs):
    """
    Calculate the dimension of the multiple actions of the env.
    env actions are assumed to be MultiDiscrete, meaning action is multionehot

    Args:
        env_spec: Environment spec (from env.specs)

    Returns:
        Sequence[int]: dimensions of the actions vectors
    """
    # Handle if passed full env spec or just observation spec
    act_spec = env_specs["input_spec"]["full_action_spec"]["action"]
    print(f"{act_spec.space}")

    return [box.n for box in act_spec.space]

get_action_vector_dims(env.specs)
```

```python
env_spec = env.specs
action_spec = env_spec.get("action", env_spec) if hasattr(env_spec, 'get') else env_spec
    
# For MultiDiscrete, extract the number of categories for each discrete action
if hasattr(action_spec, 'space') and hasattr(action_spec.space, 'boxes'):
    # MultiOneHot or MultiDiscrete with BoxList
    action_dims = [box.n for box in action_spec.space.boxes]
elif hasattr(action_spec, 'nvec'):
    # Standard MultiDiscrete
    action_dims = action_spec.nvec.tolist()
else:
    raise ValueError(f"Unsupported action spec type: {type(action_spec)}")
```
